```python
# Load libraries
import pandas as pd
import scipy.stats as st
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1

# (5 pts.) Reshape dataset election_train from long format to wide format. Hint: the
# reshaped dataset should contain 1205 rows and 6 columns.

# Load dataset and display the first five rows
data_elec = pd.read_csv('election_train.csv')
data_elec = pd.pivot_table(data_elec, index = ['Year', 'State', 'County', 'Office'], columns = 'Party', values='Votes').reset_index()
print(data_elec)
```

```python
# 2

# (20 pts.) Merge reshaped dataset election_train with dataset demographics_train. Make
# sure that you address all inconsistencies in the names of the states and the counties
# before merging. Hint: the merged dataset should contain 1200 rows.

data_demo = pd.read_csv('demographics_train.csv')
# replace state name to abbreviation
data_demo['State'] = data_demo['State'].map(us_state_abbrev)

#delete ' County'
data_elec['County'] = data_elec['County'].str.replace('County','')

# delete space and change to lowercase
data_elec['County'] = data_elec['County'].str.lower().str.strip()
data_demo['County'] = data_demo['County'].str.lower().str.strip()

data_demo['County'] = data_demo['County'].str.replace(' ._', '')
data_elec['County'] = data_elec['County'].str.replace(' ._', '')

data_merge = pd.merge(data_elec, data_demo, how='inner', on=['State', 'County'])
print(data_merge)
```

# 3

(5 pts.) Explore the merged dataset. How many variables does the dataset have? What
is the type of these variables? Are there any irrelevant or redundant variables? If so,
how will you deal with these variables?

```python
# 1200 variables.
# Yes, Citizen Voting-Age Population are many 0s, I would drop Citizen Voting-Age colume. and fill the 0 variables

print(data_merge.dtypes)
print(data_merge.isin([0]).sum())
data_merge = data_merge.drop(['Citizen Voting-Age Population', 'Year', 'Office'], 1)
print(data_merge.head)
```

# 4

(10 pts.) Search the merged dataset for missing values. Are there any missing values?
If so, how will you deal with these values?

```python
data_merge.fillna(0,inplace=True)
print(data_merge.isnull().sum())
```

# 5

(5 pts.) Create a new variable named ‚ÄúParty‚Äù that labels each county as Democratic or
Republican. This new variable should be equal to 1 if there were more votes cast for the
Democratic party than the Republican party in that county and it should be equal to 0
otherwise.

```python
data_merge['Party'] = data_merge.apply(lambda x:1 if x.Democratic > x.Republican else 0, axis=1)
print(data_merge.head)
```

# 6

(10 pts.) Compute the mean median household income for Democratic counties and
Republican counties. Which one is higher? Perform a hypothesis test to determine
whether this difference is statistically significant at the ùú∂ = ùüé. ùüéùüì significance level. What
is the result of the test? What conclusion do you make from this result?

```python
print(data_merge.groupby('Party', as_index=False)['Median Household Income'].mean())
[statistic, pvalue] = st.ttest_ind(data_merge.loc[data_merge['Party'] == 0, 'Median Household Income'],
                                   data_merge.loc[data_merge['Party'] == 1, 'Median Household Income'],
                                   equal_var = False)
print(pvalue)

# p value smaller than 0.05, so reject
```

# 7

(10 pts.) Compute the mean population for Democratic counties and Republican
counties. Which one is higher? Perform a hypothesis test to determine whether this
difference is statistically significant at the ùú∂ = ùüé. ùüéùüì significance level. What is the result
of the test? What conclusion do you make from this result?

```python
print(data_merge.groupby('Party', as_index=False)['Total Population'].mean())
[statistic, pvalue] = st.ttest_ind(data_merge.loc[data_merge['Party'] == 0, 'Total Population'],
                                   data_merge.loc[data_merge['Party'] == 1, 'Total Population'],
                                   equal_var = False)
print(pvalue)
```

# 8

(20 pts.) Compare Democratic counties and Republican counties in terms of age, gender,
race and ethnicity, and education by computing descriptive statistics and creating plots
to visualize the results. What conclusions do you make for each variable from the
descriptive statistics and the plots?

```python

# Age
data_age = data_merge[['Percent Age 29 and Under', 'Percent Age 65 and Older', 'Party']]
data_age = data_age.melt('Party', var_name='Age Range', value_name='Population')
# Gender
data_merge['Percent Male'] = data_merge.apply(lambda x: 100-x['Percent Female'], axis=1)
data_gender = data_merge[['Party', 'Percent Female', 'Percent Male']]
data_gender = data_gender.melt('Party', var_name='Gender', value_name='Population')
# Race
data_race = data_merge[['Party', 'Percent White, not Hispanic or Latino', 'Percent Black, not Hispanic or Latino', 'Percent Hispanic or Latino']]
data_race = data_race.melt('Party', var_name='Race', value_name='Population')


fig, axes = plt.subplots(1, 4, figsize = (15, 5))
sns.barplot(x='Party', y='Population', hue='Age Range', data=data_age, ax = axes[0])
sns.barplot(x='Party', y='Population', hue='Gender', data=data_gender, ax = axes[1])
sns.barplot(x='Party', y='Population', hue='Race', data=data_race, ax = axes[2])
plt.tight_layout()
```
